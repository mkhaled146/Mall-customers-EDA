### visualize data and get conclusions

```python
plt.title(" CustomerID ")
sns.countplot('CustomerID',data=df)
```

```python
df['CustomerID'].value_counts()
```

```python
plt.title(" Age")
sns.countplot('Age',data=df)
```

```python
df['Age'].value_counts()
```

```python
plt.figure(figsize=(18,8))
plt.title(" Annual Income")
sns.countplot('Annual Income (k$)',data=df)
```

```python
df['Annual Income (k$)'].value_counts()
```

```python
plt.figure(figsize=(18,8))
plt.title(" Spending Score")
sns.countplot('Spending Score (1-100)',data=df)
```

```python
df['Spending Score (1-100)'].value_counts()
```

```python
plt.title('Age vs Annual Income')
sns.regplot(x='Age',y='Annual Income (k$)',data=df)
```

```python
plt.title('Age vs Spending Score ')
sns.regplot(x='Age',y='Spending Score (1-100)',data=df)
```

```python
plt.title('Annual Income vs Spending Score ')
sns.regplot(x='Annual Income (k$)',y='Spending Score (1-100)',data=df)
```

```python
df['Age'].value_counts().plot.pie()
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(df.corr(),annot=True)
```
